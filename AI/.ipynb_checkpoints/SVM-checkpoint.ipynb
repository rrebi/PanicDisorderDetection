```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, mean_squared_error

df_train = pd.read_csv('panic_disorder_dataset_training.csv')
df_test = pd.read_csv('panic_disorder_dataset_testing.csv')

# List of columns to impute, in order of least to most missing values
columns_to_impute = ['Psychiatric History', 'Medical History', 'Substance Use']

# Function to impute a column using Random Forest
def impute_column(df, target_column, features):
    # Split the data into train and test based on missing values in the target column
    train_data = df[df[target_column].notnull()]
    test_data = df[df[target_column].isnull()]

    # Define X (features) and y (target) for training
    X_train = train_data[features]
    y_train = train_data[target_column]
    X_test = test_data[features]

    # Encode categorical variables
    X_train = pd.get_dummies(X_train, drop_first=True)
    X_test = pd.get_dummies(X_test, drop_first=True)

    # Ensure the test set has the same dummy variables as the train set
    X_test = X_test.reindex(columns = X_train.columns, fill_value=0)

    # Choose appropriate model type
    if df[target_column].dtype == 'object':
        model = RandomForestClassifier()
    else:
        model = RandomForestRegressor()

    # Train the model
    model.fit(X_train, y_train)

    # Predict missing values
    predictions = model.predict(X_test)

    # Fill the missing values in the original dataframe
    df.loc[df[target_column].isnull(), target_column] = predictions

    return df

# List of features to use for imputation (all columns except the target and Participant ID)
features_train = [col for col in df_train.columns if col not in columns_to_impute + ['Participant ID']]

# Sequentially impute each column
for column in columns_to_impute:
    df_train = impute_column(df_train, column, features_train)


# List of features to use for imputation (all columns except the target and Participant ID)
features_test = [col for col in df_test.columns if col not in columns_to_impute + ['Participant ID']]

# Sequentially impute each column
for column in columns_to_impute:
    df_test = impute_column(df_test, column, features_test)


# Verify that there are no missing values left in the imputed columns
print(df_test[columns_to_impute].isnull().sum())
print(df_train[columns_to_impute].isnull().sum())
```

```python
from sklearn.preprocessing import StandardScaler
# Drop Participant ID as it does not affect prediction
#df.drop(columns=['Participant ID'], inplace=True)

# List of numerical features to scale
numerical_features = ['Age']

# List of categorical features to encode
categorical_features = ['Gender', 'Family History', 'Personal History', 'Current Stressors', 
                        'Symptoms', 'Severity', 'Impact on Life', 'Demographics', 
                        'Medical History', 'Psychiatric History', 'Substance Use', 
                        'Coping Mechanisms', 'Social Support', 'Lifestyle Factors']

# Scale numerical features
scaler = StandardScaler()
df_train[numerical_features] = scaler.fit_transform(df_train[numerical_features])

# Encode categorical features
df_train = pd.get_dummies(df_train, columns=categorical_features, drop_first=True)

# Scale numerical features
scaler = StandardScaler()
df_test[numerical_features] = scaler.fit_transform(df_test[numerical_features])

# Encode categorical features
df_test = pd.get_dummies(df_test, columns=categorical_features, drop_first=True)

# Verify the scaled and encoded data
df_train.sample(2)
df_test.sample(2)

# Convert all boolean columns to integers
boolean_columns = df_train.select_dtypes(include=['bool']).columns
df_train[boolean_columns] = df_train[boolean_columns].astype(int)

# Convert all boolean columns to integers
boolean_columns = df_test.select_dtypes(include=['bool']).columns
df_test[boolean_columns] = df_test[boolean_columns].astype(int)

df_train.sample(2)
df_test.sample(2)
```

```python

from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
import matplotlib.pyplot as plt
import seaborn as sns

# Define features and target for training data
X_train_preprocessed = df_train.drop(columns=['Panic Disorder Diagnosis'])
train_PPD = df_train['Panic Disorder Diagnosis']
X_test_preprocessed = df_test.drop(columns=['Panic Disorder Diagnosis'])
test_PPD = df_test['Panic Disorder Diagnosis']

X_train, X_validation, Y_train, Y_validation = train_test_split(
    X_train_preprocessed,
    train_PPD,
    test_size=0.3,
    random_state=42
)

print("Training size: ", X_train.shape[0])
print("Validation size: ", X_validation.shape[0])
print("Testing size: ", X_test_preprocessed.shape[0])

```
